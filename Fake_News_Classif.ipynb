```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/train.csv')
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.reset_index(inplace=True)
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Initialize NLTK components
stop_words = set(stopwords.words('english'))
porter_stemmer = PorterStemmer()

# Assuming 'data' is your dataset and 'text_column' is the column containing text
processed_texts = []
for text in df['text']:
    words = nltk.word_tokenize(text)
    processed_words = [porter_stemmer.stem(word) for word in words if word.lower() not in stop_words]
    processed_text = ' '.join(processed_words)
    processed_texts.append(processed_text)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=1000,ngram_range=(1,3))
X=tfidf.fit_transform(processed_texts).toarray()
Y=df['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
dfc = pd.DataFrame(X_train, columns=tfidf.get_feature_names_out())
```

```python
dfc
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
```

```python
from sklearn import metrics
classifier.fit(X_train, y_train)
pred = classifier.predict(X_test)
score = metrics.accuracy_score(y_test, pred)
print(score)
```
