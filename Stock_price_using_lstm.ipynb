```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/AAPL.csv')
# df=df['Adj Close']
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['Adj Close1'] = scaler.fit_transform(df[['Adj Close']])
df=df['Adj Close1']
```

```python
X_train=df[:int(0.7*len(df))]
X_test=df[int(0.7*len(df)):]
Xn = []
Yn=[]
for i in range(len(X_train) - 5):
    current_values = X_train.iloc[i:i+5].values
    next_value = X_train.iloc[i+5]
    new_row = {'Pd1': current_values[0],
               'Pd2': current_values[1],
               'Pd3': current_values[2],
               'Pd4': current_values[3],
               'Pd5': current_values[4],}
    new1={'Nd': next_value}

    Xn.append(new_row)
    Yn.append(new1)

X_train = pd.DataFrame(Xn)
Y_train = pd.DataFrame(Yn)

Y_train
```

```python
Xn = []
Yn=[]
for i in range(len(X_test) - 5):
    current_values = X_test.iloc[i:i+5].values
    next_value = X_test.iloc[i+5]
    new_row = {'Pd1': current_values[0],
               'Pd2': current_values[1],
               'Pd3': current_values[2],
               'Pd4': current_values[3],
               'Pd5': current_values[4],}
    new1={'Nd': next_value}

    Xn.append(new_row)
    Yn.append(new1)

X_test = pd.DataFrame(Xn)
Y_test = pd.DataFrame(Yn)
Y_test
```

```python
X_train=np.array(X_train)
X_train=X_train.reshape(X_train.shape[0],X_train.shape[1],1)
Y_train=np.array(Y_train)
X_test=np.array(X_test)
X_test=X_test.reshape(X_test.shape[0],X_test.shape[1],1)
Y_test_test=np.array(Y_test)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
model = Sequential()
model.add(LSTM(32, return_sequences=True, input_shape=(5, 1)))
model.add(LSTM(16))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

```

```python
model.fit(X_train,Y_train,epochs=100,)
```

```python
X_pred=model.predict(X_train)
X_predt=model.predict(X_test)
```

```python
X_pred=scaler.inverse_transform(X_pred)
X_predt=scaler.inverse_transform(X_predt)
```

```python
from sklearn.metrics import mean_squared_error
terror=np.sqrt(mean_squared_error(Y_train,X_pred))
terror
```

```python
terror=np.sqrt(mean_squared_error(Y_test,X_pred))
terror
```
